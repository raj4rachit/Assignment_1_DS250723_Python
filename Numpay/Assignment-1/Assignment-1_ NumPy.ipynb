```python
import numpy as np
```

```python
# 1. Create a null vector of size 10 but the fifth value which is 1.
null_vector = np.zeros(10)
null_vector[4] = 1
print("Null vector with 5th value as 1:", null_vector)
```

```python
# 2. Create a vector with values ranging from 10 to 49.
vector_10_to_49 = np.arange(10, 50)
print("Vector from 10 to 49:", vector_10_to_49)
```

```python
# 3. Create a 3x3 matrix with values ranging from 0 to 8
matrix_3x3 = np.arange(9).reshape(3, 3)
print("3x3 matrix with values from 0 to 8:\n", matrix_3x3)
```

```python
# 4. Find indices of non-zero elements from [1,2,0,0,4,0]
arr = np.array([1, 2, 0, 0, 4, 0])
nonzero_indices = np.nonzero(arr)
print("Indices of non-zero elements:", nonzero_indices)
```

```python
# 5. Create a 10x10 array with random values and find the minimum and maximum values.
random_array_10x10 = np.random.rand(10, 10)
min_value = np.min(random_array_10x10)
max_value = np.max(random_array_10x10)
print("Minimum value:", min_value)
print("Maximum value:", max_value)
```

```python
# 6. Create a random vector of size 30 and find the mean value.
random_vector_30 = np.random.rand(30)
mean_value = np.mean(random_vector_30)
print("Mean value of the random vector:", mean_value)
```
