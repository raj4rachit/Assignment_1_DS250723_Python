```python
# 1=> Implement Binary tree

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursively(self.root, data)

    def _insert_recursively(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursively(node.left, data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursively(node.right, data)

    def search(self, data):
        return self._search_recursively(self.root, data)

    def _search_recursively(self, node, data):
        if node is None:
            return False
        if node.data == data:
            return True
        if data < node.data:
            return self._search_recursively(node.left, data)
        else:
            return self._search_recursively(node.right, data)

    def inorder_traversal(self):
        result = []
        self._inorder_traversal(self.root, result)
        return result

    def _inorder_traversal(self, node, result):
        if node:
            self._inorder_traversal(node.left, result)
            result.append(node.data)
            self._inorder_traversal(node.right, result)

# Example usage with user input:
if __name__ == "__main__":
    tree = BinaryTree()
    
    while True:
        user_input = input("Enter an integer to insert into the binary tree (or 'q' to quit): ")
        if user_input == 'q':
            break
        try:
            value = int(user_input)
            tree.insert(value)
        except ValueError:
            print("Invalid input. Please enter an integer or 'q' to quit.")

    print("Inorder Traversal:", tree.inorder_traversal())

    search_value = input("Enter a value to search for in the tree: ")
    try:
        search_value = int(search_value)
        if tree.search(search_value):
            print(f"{search_value} is found in the tree.")
        else:
            print(f"{search_value} is not found in the tree.")
    except ValueError:
        print("Invalid input. Please enter an integer for searching.")
```

```python
# 5=> Implement BFS (Breath First Search) and DFS (Depth First Search)

from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursively(self.root, data)

    def _insert_recursively(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursively(node.left, data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursively(node.right, data)

    def bfs(self):
        result = []
        if self.root is None:
            return result

        queue = deque()
        queue.append(self.root)

        while queue:
            node = queue.popleft()
            result.append(node.data)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        return result

    def dfs_preorder(self):
        result = []
        self._dfs_preorder_recursive(self.root, result)
        return result

    def _dfs_preorder_recursive(self, node, result):
        if node is not None:
            result.append(node.data)
            self._dfs_preorder_recursive(node.left, result)
            self._dfs_preorder_recursive(node.right, result)

    def dfs_inorder(self):
        result = []
        self._dfs_inorder_recursive(self.root, result)
        return result

    def _dfs_inorder_recursive(self, node, result):
        if node is not None:
            self._dfs_inorder_recursive(node.left, result)
            result.append(node.data)
            self._dfs_inorder_recursive(node.right, result)

    def dfs_postorder(self):
        result = []
        self._dfs_postorder_recursive(self.root, result)
        return result

    def _dfs_postorder_recursive(self, node, result):
        if node is not None:
            self._dfs_postorder_recursive(node.left, result)
            self._dfs_postorder_recursive(node.right, result)
            result.append(node.data)

# Example usage:
if __name__ == "__main__":
    tree = BinaryTree()
    
    while True:
        user_input = input("Enter an integer to insert into the binary tree (or 'q' to quit): ")
        if user_input == 'q':
            break
        try:
            value = int(user_input)
            tree.insert(value)
        except ValueError:
            print("Invalid input. Please enter an integer or 'q' to quit.")

    print("Breadth-First Search (BFS):", tree.bfs())
    print("Depth-First Search (DFS) - Preorder:", tree.dfs_preorder())
    print("Depth-First Search (DFS) - Inorder:", tree.dfs_inorder())
    print("Depth-First Search (DFS) - Postorder:", tree.dfs_postorder())
```

```python
# 6=> Find sum of all left leaves in a given Binary Tree

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursively(self.root, data)

    def _insert_recursively(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursively(node.left, data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursively(node.right, data)

    def sum_of_left_leaves(self):
        return self._sum_of_left_leaves_recursive(self.root, is_left=False)

    def _sum_of_left_leaves_recursive(self, node, is_left=False):
        if node is None:
            return 0
        
        if node.left is None and node.right is None and is_left:
            return node.data
        
        left_sum = self._sum_of_left_leaves_recursive(node.left, is_left=True)
        right_sum = self._sum_of_left_leaves_recursive(node.right, is_left=False)
        
        return left_sum + right_sum

# Example usage:
if __name__ == "__main__":
    tree = BinaryTree()
    
    while True:
        user_input = input("Enter an integer to insert into the binary tree (or 'q' to quit): ")
        if user_input == 'q':
            break
        try:
            value = int(user_input)
            tree.insert(value)
        except ValueError:
            print("Invalid input. Please enter an integer or 'q' to quit.")

    left_leaves_sum = tree.sum_of_left_leaves()
    print("Sum of left leaves in the tree:", left_leaves_sum)
```

```python
# 7=> Find sum of all nodes of the given perfect binary tree

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root_data):
        self.root = Node(root_data)

    def insert(self, data):
        self._insert_recursively(self.root, data)

    def _insert_recursively(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursively(node.left, data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursively(node.right, data)

    def height(self, node):
        if node is None:
            return 0
        else:
            left_height = self.height(node.left)
            right_height = self.height(node.right)
            return max(left_height, right_height) + 1

    def sum_of_all_nodes(self):
        return self._sum_of_all_nodes_recursive(self.root)

    def _sum_of_all_nodes_recursive(self, node):
        if node is None:
            return 0
        return node.data + self._sum_of_all_nodes_recursive(node.left) + self._sum_of_all_nodes_recursive(node.right)

# Example usage:
if __name__ == "__main__":
    tree = BinaryTree(1)
    
    for i in range(2, 16):
        tree.insert(i)

    total_nodes = (2 ** tree.height(tree.root)) - 1
    sum_of_all = tree.sum_of_all_nodes()

    print("Total number of nodes in the perfect binary tree:", total_nodes)
    print("Sum of all nodes in the perfect binary tree:", sum_of_all)
```

```python
# 9=> Find maximum level sum in Binary Tree

from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursively(self.root, data)

    def _insert_recursively(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursively(node.left, data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursively(node.right, data)

    def max_level_sum(self):
        if self.root is None:
            return 0

        max_sum = float('-inf')
        max_level = 1
        level = 1

        queue = deque()
        queue.append(self.root)

        while queue:
            level_sum = 0
            level_size = len(queue)

            for i in range(level_size):
                node = queue.popleft()
                level_sum += node.data

                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)

            if level_sum > max_sum:
                max_sum = level_sum
                max_level = level

            level += 1

        return max_level, max_sum

# Example usage:
if __name__ == "__main__":
    tree = BinaryTree()
    
    while True:
        user_input = input("Enter an integer to insert into the binary tree (or 'q' to quit): ")
        if user_input == 'q':
            break
        try:
            value = int(user_input)
            tree.insert(value)
        except ValueError:
            print("Invalid input. Please enter an integer or 'q' to quit.")

    max_level, max_sum = tree.max_level_sum()

    print("Maximum level sum is at level", max_level, "with a sum of", max_sum)
```
